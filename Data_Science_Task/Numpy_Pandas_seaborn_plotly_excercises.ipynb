# **Numpy Excercises**

**Excercise-1**

Task: Modify the code to create a 4x4 matrix of random integers between 10 and 50, then set all
values greater than 40 to 0.

```python

import numpy as np 
array=np.random.randint(10,50,(4,4))
print("Original Array\n",array)
array=np.where(array>40,0,array)
print("\n Modified Array \n",array)
```

**Excercise-2**

Task: Modify the code to compute the median and standard deviation instead of the mean, and
apply it to a 1D array of 20 random floats between 0 and 10.

```python
array=np.random.rand(1,20) * 10
median=np.median(array)
Std=np.std(array)
print(f" Median={median} and Standar Deviation={Std}")
```

# **Pandas Exercises**

**Excercise-3**

**Task: Modify the code to filter rows where age is greater than 25 and city is "New York".**

```python
import pandas as pd

# Original code
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, 35, 20],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago']
}

df = pd.DataFrame(data)
filtered_df = df[(df['age'] > 25) & (df['city']=='New York')]
print("Filtered DataFrame:\n", filtered_df)
```

**Exercise 4: Grouping and Aggregation**

Task: Modify the code to group by city and compute the maximum age for each city.

```python
import pandas as pd
# Original code
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 30, 35, 20, 28],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angel']
}
df = pd.DataFrame(data)
grouped = df.groupby('city')['age'].max().reset_index()
print("Grouped means:\n", grouped)
```

# **Seaborn Exercises**

**Exercise 5: Boxplot Customization**

Task: Modify the code to create a boxplot of total_bill grouped by day, with points colored by
sex.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Original code
tips = sns.load_dataset('tips')
sns.boxplot(data=tips, 
            x='day',
            y='total_bill',
            hue='sex'
           )
plt.show()
```

**Exercise 6: Correlation Heatmap** 

Task: Modify the code to create a heatmap of the correlation matrix for numerical columns in the
tips dataset, with annotations.

```python
corr_matrix = tips.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Columns in Tips Dataset")
plt.show()
```

# **Plotly Excercises**
**Excercise-7**

**Task: Modify the code to create a scatter plot of total_bill vs. tip, with points colored by day and
sized by size.**

```python
import plotly.express as px
import seaborn as sns
tips=sns.load_dataset('Tips')

plot=px.scatter(
    data_frame=tips,
    x='total_bill',
    y='tip',
    color='day',
    size='size'
)
plot.show()
```
